```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, MinMaxScaler, OneHotEncoder, LabelEncoder
from sklearn.feature_selection import SelectKBest, chi2
from sklearn.decomposition import PCA
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
```

```python
df = pd.read_csv('smartphone-uncleaned.csv')
```

```python
numeric_df = df.select_dtypes(include=[np.number])

standardized_df = (numeric_df - numeric_df.mean()) / numeric_df.std()

final_df = df.copy()
final_df[numeric_df.columns] = standardized_df

print("\nStandardized Data:")
print(final_df)

standardized_file_path = 'standardized_data.csv' 
final_df.to_csv(standardized_file_path, index=False)

print(f"\nStandardized data saved to {standardized_file_path}")
```

```python
numeric_df = df.select_dtypes(include=[np.number])

normalized_df = (numeric_df - numeric_df.min()) / (numeric_df.max() - numeric_df.min())

final_df = df.copy()
final_df[numeric_df.columns] = normalized_df

print("\nNormalized Data:")
print(final_df)

normalized_file_path = 'normalized_data.csv'  
final_df.to_csv(normalized_file_path, index=False)

print(f"\nNormalized data saved to {normalized_file_path}")
```
